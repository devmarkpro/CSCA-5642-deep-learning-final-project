```python
import pandas as pd
```

```python
captions = pd.read_csv("./data/captions.txt")
captions.astype(str)
```

```python
captions["image"].astype(str).str.replace(r"\s+", " ", regex=True).str.strip(' "\'')
captions["caption"].astype(str).str.replace(r"\s+", " ", regex=True).str.strip(' "\'')
```
